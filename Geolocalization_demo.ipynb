# Welcome to pyiunstir!

This notebook shows how to use the geolocalization information to analyze the availabel corpus

```python
import os 
import json

import matplotlib.pyplot as plt
os.environ['PROJ_LIB'] = '$HOME/miniconda3/share/proj'
from mpl_toolkits.basemap import Basemap

from pyiunstir.encoding import *

# Load the encodings
f_json = open('corpus/iberian.json', 'r')
symbols = json.load(f_json)
f_json.close()

# Load a collection of text written using the Northeastern script
f_json = open('corpus/NE_database.json', 'r')
corpus_NE = json.load(f_json)
f_json.close()

%matplotlib inline
```

## Find coins labelled with the genitive plural ending "-ken", meaning "from the people of" 

```python
qx = []
qy = []
for scp in corpus_NE:
    flag = 0
    if scp['support'] == 'coin':
        for t in scp['text_simplified']:
            if t.endswith('ken'):
                flag = 1
        if flag == 1:
            qx.append(float(scp['x']))
            qy.append(float(scp['y']))
            #print(scp['text_simplified'])
            
fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution='h', 
            lat_0=40, lon_0=-0.6,
            width=2E6, height=1.2E6)

m.shadedrelief()
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')
m.drawstates(color='gray')

# 2. scatter city data, with color reflecting population
# and size reflecting area
m.scatter(qy, qx, latlon=True,
          c=[2] * len(qx));
plt.title('Distribution of -ken coins');
```

# Now, let's look for steles with the seltar/siltar word on it

```python
qx = []
qy = []
for scp in corpus_NE:
        flag = 0
        for t in scp['text_simplified']:
            if t.find('seltar') >= 0 or t.find('siltar') >= 0 :
                flag = 1
        if flag == 1:
            qx.append(float(scp['x']))
            qy.append(float(scp['y']))
            print(scp['text_simplified'])

fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution='h', 
            lat_0=40, lon_0=-0.6,
            width=2E6, height=1.2E6)

m.shadedrelief()
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')
m.drawstates(color='gray')

# 2. scatter city data, with color reflecting population
# and size reflecting area
m.scatter(qy, qx, latlon=True,
          c=[2] * len(qx));
plt.title('Tombstones with seltar');
```

# Find texts with "iunstir" greetings

```python
qx = []
qy = []
for scp in corpus_NE:
        flag = 0
        for t in scp['text_simplified']:
            if t.find('iunstir') >= 0:
                flag = 1
        if flag == 1:
            qx.append(float(scp['x']))
            qy.append(float(scp['y']))

fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution='h', 
            lat_0=40, lon_0=-0.6,
            width=2E6, height=1.2E6)

m.shadedrelief()
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')
m.drawstates(color='gray')

# 2. scatter city data, with color reflecting population
# and size reflecting area
m.scatter(qy, qx, latlon=True,
          c=[2] * len(qx));
plt.title('iunstir distribution');
```
