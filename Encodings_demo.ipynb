# Welcome to pyiunstir!

You can start using the iberian encodings with just:

```python
from pyiunstir.encoding import *
import json
import ast

f_json = open('corpus/iberian.json', 'r')
symbols = json.load(f_json)
f_json.close()
```

### Encoding and decoding is carried out by the encode_iberian and decode_iberian methods
### Each script "supports" in a different manner "foreign words" such as personal names or places
### Let's try a "Hello, world!" using  all available scripts: 

## neitin:iunstir:alvarode:egiar:madridku 
### Roughly meaning Greetings. Made by Alvaro from Madrid.

### First, using the northeastern signary, the most popular one

```python
from IPython.display import SVG
%matplotlib inline

text = 'neitin:iunstir:alvarode:egiar:madridku'

welcome_NE = encode_iberian(text, symbols, signary='northeastern' )
w_welcome_NE_iberian = decode_iberian(welcome_NE, symbols, 'iberian')
SVG(render_letter(w_welcome_NE_iberian))
```

### Now with a greek iberian alphabet

```python
welcome_GI = encode_iberian(text, symbols, signary='greekiberian' )
w_welcome_GI_iberian = decode_iberian(welcome_GI, symbols, 'iberian')
SVG(render_letter(w_welcome_GI_iberian))
```

### Using the southeastern script, which needs to be inverted to be read from right to left, is also possible

```python
welcome_SE = encode_iberian(text, symbols, signary='southeastern' )
w_welcome_SE_iberian = decode_iberian(welcome_SE, symbols, 'iberian')
SVG(render_letter(list(reversed(w_welcome_SE_iberian))))
```

### And finally, back to extended ASCII (simplified, note the g --> k and d-->t)

```python
decode_iberian(welcome_NE, symbols, 'simplified')
```

### ....and dual (please note the change in the original Madridku to madirdeku)

```python
decode_iberian(welcome_NE, symbols, 'dual')
```

### ...and from Greek-iberian (note the change for m --> nb, but not for dku)

```python
decode_iberian(welcome_GI, symbols, 'dual')
```
